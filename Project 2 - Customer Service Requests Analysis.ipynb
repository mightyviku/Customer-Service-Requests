```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
# Read the dataset into a dataframe and visualize the dataset
df = pd.read_csv('311_Service_Requests_from_2010_to_Present.csv')
df.sample(10)
```

```python
# Print the column names and shape of the dataframe
print(df.columns)
print(df.shape)
```

```python
# Identify the number of null values for each column
nan = df.isna().sum()
nan
```

```python
nan/df.shape[0]*100
```

```python
type(df.isna().sum())
df.shape[0]*30/100
```

```python
# Plotting count of missing values for each feature
plt.figure(figsize=(12,12))
plt.xlabel('Missing Values')
sns.barplot(x=nan.values, y=nan.index)
```

```python
# Drop columns with >30% of values missing
df.dropna(axis='columns', thresh=109367, inplace=True)
df.isna().sum()
```

```python
df.isna().sum()/df.shape[0]*100
```

```python
# Drop rows with any remaining missing values
df.dropna(inplace=True)
df.shape
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
# Check the datatypes of all the columns
df.info()
```

```python
# Convert the Created Date and Closed Date columns to datetime format
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
df.info()
```

```python
# Use the converted dates to calculate the Time Elapsed between creation and closing of each case
df['Time Elapsed'] = df['Closed Date'] - df['Created Date']
df['Time Elapsed'].head()
```

```python
# Check the basic stats of Time Elapsed to ensure no strange values (negative time)
df['Time Elapsed'].describe()
```

```python
df.head()
```

```python
# Create a new column that shows the time elapsed purely in seconds
df['Time Elapsed (sec)'] = df['Time Elapsed'].dt.total_seconds()
df['Time Elapsed (sec)'].head()
```

```python
df['Time Elapsed (sec)'].describe()
```

```python
# Plotting the number of complaints in the data by city
plt.figure(figsize=(16,12))
sns.countplot(data=df, y='City')
```

```python
df[df.City=='BROOKLYN'].head()
```

```python
# Counting the number of complaints in the data by complaint type
plt.figure(figsize=(16,12))
sns.countplot(data=df, y='Complaint Type')
```

```python
# Creating a hexbin plot to visualize the concentration of complaints across Brooklyn
plt.figure(figsize=(16,12))
sns.jointplot(data=df[df.City=='BROOKLYN'], y='Latitude', x='Longitude', kind='hex')
```
